```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import re
import pandas as pd
import os
```

```python
def get_movie_list():
    driver = webdriver.Chrome()
    driver.get("http://www.hindilyrics.net/hindi-songs-translations.html")
    link_list = driver.find_elements_by_tag_name('li')
    movie_link_list = []
    for link in link_list:
        movie_link_list.append(link.find_element_by_tag_name('a').get_attribute('href'))
    return movie_link_list
```

```python
def get_song_list(url):
    driver.get(url)
    link_list = driver.find_elements_by_tag_name('li')
    song_list = []
    for link in link_list:
        song_list.append(link.find_element_by_tag_name('a').get_attribute('href'))
    return song_list
```

```python
def get_lyrics_of_song(url):
    driver.get(url)
    lyrics = driver.find_elements_by_id('t')
    song_lyrics = []
    for ly in lyrics:
        song_lyrics.append(ly.text)
    return song_lyrics
```

```python
def get_all():
    movie_list = get_movie_list()
    df = pd.DataFrame(columns=['url', 'lyrics'])
    for movie in movie_list:
        print(movie)
        song_list = get_song_list(movie)
        for song in song_list:
            song_lyrics = get_lyrics_of_song(song)
            df.loc[len(df)] = [song, song_lyrics]
    return df
```

```python
df = get_all()
df.to_csv("song_bolly_translated.csv")
```
